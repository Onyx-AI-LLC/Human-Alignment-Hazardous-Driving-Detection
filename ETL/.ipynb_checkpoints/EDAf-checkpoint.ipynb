```python
# screen dimensions exploratory data analysis
# this notebook analyzes the screen dimensions used by participants in the driving simulation

import pandas as pd
import ast
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("starting screen dimensions eda...")

# load the survey data
print("loading survey data...")
df = pd.read_csv('../data/raw/survey_results_raw.csv')
print(f"loaded {len(df)} records")

# display basic info about the dataset
print("\ndataset info:")
print(f"columns: {list(df.columns)}")
print(f"shape: {df.shape}")

# show sample window dimensions
print("\nsample window dimensions entries:")
for i in range(min(5, len(df))):
    print(f"  {df.iloc[i]['windowDimensions']}")

# extract dimensions from the windowDimensions column
print("\nextracting window dimensions...")
dimensions = []

for idx, row in df.iterrows():
    try:
        # parse the string representation of the dictionary
        window_dim_str = row['windowDimensions']
        
        # safely evaluate the string as a python literal
        window_dim_dict = ast.literal_eval(window_dim_str)
        
        width = window_dim_dict['width']
        height = window_dim_dict['height']
        
        dimensions.append({
            'record_id': row['_id'],
            'user_id': row['userId'],
            'video_id': row['videoId'],
            'width': width,
            'height': height,
            'aspect_ratio': round(width/height, 3)
        })
        
    except Exception as e:
        print(f"error parsing row {idx}: {e}")
        continue

dimensions_df = pd.DataFrame(dimensions)
print(f"successfully extracted {len(dimensions_df)} dimension records")

# basic analysis
print("\n" + "="*60)
print("screen dimensions analysis")
print("="*60)

# unique dimensions
unique_dims = dimensions_df[['width', 'height']].drop_duplicates()
print(f"number of unique screen dimensions: {len(unique_dims)}")

print("\nall unique screen dimensions:")
dim_summary = []
for _, row in unique_dims.iterrows():
    count = len(dimensions_df[(dimensions_df['width'] == row['width']) & 
                             (dimensions_df['height'] == row['height'])])
    aspect_ratio = round(row['width']/row['height'], 3)
    dim_summary.append({
        'dimensions': f"{row['width']}x{row['height']}",
        'aspect_ratio': aspect_ratio,
        'count': count,
        'percentage': round(count/len(dimensions_df)*100, 1)
    })

dim_summary_df = pd.DataFrame(dim_summary).sort_values('count', ascending=False)
print(dim_summary_df.to_string(index=False))

# aspect ratio analysis
print(f"\nunique aspect ratios found: {len(dimensions_df['aspect_ratio'].unique())}")
aspect_counts = Counter(dimensions_df['aspect_ratio'])
print("aspect ratio distribution:")
for aspect, count in aspect_counts.most_common():
    percentage = count/len(dimensions_df)*100
    print(f"  {aspect}: {count} records ({percentage:.1f}%)")

# statistics
print(f"\nwidth statistics:")
print(f"  min: {dimensions_df['width'].min()}px")
print(f"  max: {dimensions_df['width'].max()}px") 
print(f"  mean: {dimensions_df['width'].mean():.1f}px")
print(f"  median: {dimensions_df['width'].median()}px")
print(f"  std: {dimensions_df['width'].std():.1f}px")

print(f"\nheight statistics:")
print(f"  min: {dimensions_df['height'].min()}px")
print(f"  max: {dimensions_df['height'].max()}px")
print(f"  mean: {dimensions_df['height'].mean():.1f}px")
print(f"  median: {dimensions_df['height'].median()}px")
print(f"  std: {dimensions_df['height'].std():.1f}px")

# create comprehensive visualizations
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# distribution of widths
axes[0, 0].hist(dimensions_df['width'], bins=15, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 0].set_title('distribution of screen widths', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('width (pixels)')
axes[0, 0].set_ylabel('frequency')
axes[0, 0].grid(True, alpha=0.3)

# distribution of heights  
axes[0, 1].hist(dimensions_df['height'], bins=15, alpha=0.7, color='lightcoral', edgecolor='black')
axes[0, 1].set_title('distribution of screen heights', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('height (pixels)')
axes[0, 1].set_ylabel('frequency')
axes[0, 1].grid(True, alpha=0.3)

# aspect ratio distribution
axes[0, 2].hist(dimensions_df['aspect_ratio'], bins=15, alpha=0.7, color='orange', edgecolor='black')
axes[0, 2].set_title('distribution of aspect ratios', fontsize=12, fontweight='bold')
axes[0, 2].set_xlabel('aspect ratio (width/height)')
axes[0, 2].set_ylabel('frequency')
axes[0, 2].grid(True, alpha=0.3)

# scatter plot of width vs height
scatter = axes[1, 0].scatter(dimensions_df['width'], dimensions_df['height'], 
                            alpha=0.6, color='green', s=50)
axes[1, 0].set_title('width vs height scatter plot', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('width (pixels)')
axes[1, 0].set_ylabel('height (pixels)')
axes[1, 0].grid(True, alpha=0.3)

# add annotations for unique dimensions
for _, row in unique_dims.iterrows():
    count = len(dimensions_df[(dimensions_df['width'] == row['width']) & 
                             (dimensions_df['height'] == row['height'])])
    if count > 1:  # only annotate if more than 1 record
        axes[1, 0].annotate(f'{int(row["width"])}x{int(row["height"])}\n({count})', 
                           (row['width'], row['height']), 
                           xytext=(5, 5), textcoords='offset points', 
                           fontsize=8, alpha=0.8)

# box plot of dimensions
box_data = [dimensions_df['width'], dimensions_df['height']]
box_plot = axes[1, 1].boxplot(box_data, labels=['width', 'height'], patch_artist=True)
box_plot['boxes'][0].set_facecolor('skyblue')
box_plot['boxes'][1].set_facecolor('lightcoral')
axes[1, 1].set_title('width and height distributions', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('pixels')
axes[1, 1].grid(True, alpha=0.3)

# dimension frequency bar chart
top_dims = dim_summary_df.head(8)  # top 8 most common
bars = axes[1, 2].bar(range(len(top_dims)), top_dims['count'], 
                     color='purple', alpha=0.7)
axes[1, 2].set_title('most common screen dimensions', fontsize=12, fontweight='bold')
axes[1, 2].set_xlabel('dimension')
axes[1, 2].set_ylabel('frequency')
axes[1, 2].set_xticks(range(len(top_dims)))
axes[1, 2].set_xticklabels(top_dims['dimensions'], rotation=45, ha='right')
axes[1, 2].grid(True, alpha=0.3)

# add count labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    axes[1, 2].text(bar.get_x() + bar.get_width()/2., height,
                   f'{int(height)}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.savefig('screen_dimensions_comprehensive_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# recommendations for standardization
print("\n" + "="*60)
print("recommendations for standardization")
print("="*60)

# find most common dimensions
most_common_dims = dim_summary_df.iloc[0]
print(f"most common screen dimensions: {most_common_dims['dimensions']}")
print(f"  - used by {most_common_dims['count']} records ({most_common_dims['percentage']}%)")
print(f"  - aspect ratio: {most_common_dims['aspect_ratio']}")

# find most common aspect ratio
most_common_aspect = aspect_counts.most_common(1)[0]
print(f"\nmost common aspect ratio: {most_common_aspect[0]}")
print(f"  - used by {most_common_aspect[1]} records ({most_common_aspect[1]/len(dimensions_df)*100:.1f}%)")

# calculate statistics for standardization decision
records_needing_resize = len(dimensions_df) - most_common_dims['count']
print(f"\nif standardizing to most common dimensions ({most_common_dims['dimensions']}):")
print(f"  - records needing resize: {records_needing_resize} ({records_needing_resize/len(dimensions_df)*100:.1f}%)")
print(f"  - records already correct: {most_common_dims['count']} ({most_common_dims['percentage']}%)")

# alternative recommendations
median_width = int(dimensions_df['width'].median())
median_height = int(dimensions_df['height'].median())
median_aspect = round(median_width/median_height, 3)

print(f"\nalternative option (median-based dimensions):")
print(f"  - dimensions: {median_width}x{median_height}")
print(f"  - aspect ratio: {median_aspect}")

# check if median dimensions exist in data
median_exact_matches = len(dimensions_df[(dimensions_df['width'] == median_width) & 
                                        (dimensions_df['height'] == median_height)])
print(f"  - exact matches: {median_exact_matches}")

# final recommendation
print(f"\n{'='*30} final recommendation {'='*30}")
if most_common_dims['percentage'] >= 30:
    print(f"recommend standardizing to: {most_common_dims['dimensions']}")
    print(f"reason: most common dimensions with {most_common_dims['percentage']}% coverage")
else:
    print(f"recommend standardizing to: {median_width}x{median_height}")
    print(f"reason: median-based dimensions, good middle ground")

# save analysis results
dimensions_df.to_csv('processed_screen_dimensions.csv', index=False)
dim_summary_df.to_csv('dimension_summary.csv', index=False)

print(f"\nanalysis complete!")
print(f"saved processed data to:")
print(f"  - processed_screen_dimensions.csv ({len(dimensions_df)} records)")
print(f"  - dimension_summary.csv ({len(dim_summary_df)} unique dimensions)")
print(f"  - screen_dimensions_comprehensive_analysis.png (visualization)")

# display final summary
print(f"\n{'='*60}")
print("summary")  
print(f"{'='*60}")
print(f"total simulation records: {len(dimensions_df)}")
print(f"unique participants: {dimensions_df['user_id'].nunique()}")
print(f"unique videos: {dimensions_df['video_id'].nunique()}")
print(f"unique screen dimensions: {len(unique_dims)}")
print(f"screen size variability: {dimensions_df['width'].std():.1f}px width, {dimensions_df['height'].std():.1f}px height")
print(f"most common aspect ratio: {most_common_aspect[0]} ({most_common_aspect[1]/len(dimensions_df)*100:.1f}% of records)")
```
